## Unzip and Load Data

### Subtask:

**Reasoning**:
Unzip the `archive.zip` file to extract its contents.



````python
import zipfile
import os

# Define the path to the zip file
zip_file_path = '/content/archive.zip'

# Define the directory to extract to
extract_dir = '/content/'

# Create the extraction directory if it doesn't exist
os.makedirs(extract_dir, exist_ok=True)

# Unzip the file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"'{zip_file_path}' unzipped to '{extract_dir}' successfully.")

# List the contents of the extracted directory to verify
print("Contents of the extracted directory:")
for root, dirs, files in os.walk(extract_dir):
    for file in files:
        print(os.path.join(root, file))
````

**Reasoning**:

**Reasoning**:
To address missing values as part of data cleaning, I will first calculate the percentage of missing values for each column and display them. This will help in understanding the extent of missing data and prioritizing columns for further treatment.



````python
print("åã«ã©ã ã®æ¬ æå¤ã®å²åã¨ãã¼ã¿å:")
missing_values = pd.DataFrame({
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df)) * 100,
    'Dtype': df.dtypes
})
missing_values = missing_values[missing_values['Missing_Count'] > 0].sort_values(by='Missing_Percentage', ascending=False)
print(missing_values)
````

**Reasoning**:
Now that columns with very high missing percentages have been removed, I need to reassess the remaining missing values in the DataFrame. This will provide an updated view of the data's cleanliness and help in deciding how to handle the remaining missing data.



````python
print("ã«ã©ã åé¤å¾ã®åã«ã©ã ã®æ¬ æå¤ã®å²åã¨ãã¼ã¿å:")
missing_values_after_drop = pd.DataFrame({
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df)) * 100,
    'Dtype': df.dtypes
})
missing_values_after_drop = missing_values_after_drop[missing_values_after_drop['Missing_Count'] > 0].sort_values(by='Missing_Percentage', ascending=False)
print(missing_values_after_drop)
````

**Reasoning**:

````python
print("æçµçãªæ¬ æå¤ã®å²åã¨ãã¼ã¿å:")
final_missing_values = pd.DataFrame({
    'Missing_Count': df.isnull().sum(),
    'Missing_Percentage': (df.isnull().sum() / len(df)) * 100,
    'Dtype': df.dtypes
})
final_missing_values = final_missing_values[final_missing_values['Missing_Count'] > 0].sort_values(by='Missing_Percentage', ascending=False)
print(final_missing_values)
````

**Reasoning**:
To complete the missing value treatment, I will impute the remaining missing numerical values with their median and categorical values with their mode. This helps to retain as much data as possible while ensuring all columns are complete for further analysis.



````python
print("æ®ãã®æ¬ æå¤ãå¦çãã¾ãï¼æ°å¤ãã¼ã¿ã¯ä¸­å¤®å¤ãã«ãã´ãªãã¼ã¿ã¯æé »å¤ã§è£å®ï¼ã")

for col in final_missing_values.index:
    if final_missing_values.loc[col, 'Dtype'] in ['float64', 'int64']:
        # æ°å¤ãã¼ã¿ã¯ä¸­å¤®å¤ã§è£å®
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"{col} ã®æ¬ æå¤ãä¸­å¤®å¤ {median_val} ã§è£å®ãã¾ããã")
    else:
        # ã«ãã´ãªãã¼ã¿ã¯æé »å¤ã§è£å®
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"{col} ã®æ¬ æå¤ãæé »å¤ {mode_val} ã§è£å®ãã¾ããã")

print("æ¬ æå¤å¦çå¾ã®åã«ã©ã ã®æ¬ æå¤ã®ç·æ°:")
print(df.isnull().sum()[df.isnull().sum() > 0])
````

**Reasoning**:

````python
print("RAMã¨SSD Capacityã«ã©ã ãæ°å¤åã«ã¯ãªã¼ãã³ã°ãã¾ãã")

# RAMã«ã©ã ã®ã¯ãªã¼ãã³ã°ã¨æ°å¤åã¸ã®å¤æ
# 'TB' ã 'GB' ã«å¤æããåä½ãåé¤ãã¦æ°å¤ã«å¤æ
def convert_to_gb(value):
    value = str(value).upper().strip()
    if 'TB' in value:
        return int(float(value.replace('TB', '').strip()) * 1024)
    elif 'GB' in value:
        return int(float(value.replace('GB', '').strip()))
    else:
        return int(float(value))

df['RAM'] = df['RAM'].apply(convert_to_gb)

# SSD Capacityã«ã©ã ã®ã¯ãªã¼ãã³ã°ã¨æ°å¤åã¸ã®å¤æ
df['SSD Capacity'] = df['SSD Capacity'].apply(convert_to_gb)

print("RAMã«ã©ã ã®ã¯ãªã¼ãã³ã°å¾ã®ãã¼ã¿åã¨æåã®5ã¤ã®å¤:")
print(df['RAM'].dtype)
print(df['RAM'].head())

print("\nSSD Capacityã«ã©ã ã®ã¯ãªã¼ãã³ã°å¾ã®ãã¼ã¿åã¨æåã®5ã¤ã®å¤:")
print(df['SSD Capacity'].dtype)
print(df['SSD Capacity'].head())
````
